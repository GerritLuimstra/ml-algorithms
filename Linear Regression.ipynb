```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def get_y_by_slope(slope, base, x):
    return (slope * x) + base
```

```python
# Let's generate some example data
xs = [1, 1.5, 2, 2.2, 3, 4]
ys = [0.2, 0.3, 0.33, 0.44, 0.77, 0.99]

# Define a simple 'model' at random as an example
mod_param_1 = 0.5
mod_param_2 = 0.05

plt.scatter(xs, ys, c = "red")
plt.xlabel('Years in the education system')
plt.ylabel('Percentage of mental frustration')
plt.axis([0, 4, 0, 1])
plt.plot([0, 4], [get_y_by_slope(mod_param_1, mod_param_2, 0), get_y_by_slope(mod_param_1, mod_param_2, 4)])
plt.show()
```

```python
def linear_regression(xs, ys, lr = 0.05, epochs = 5):
    
    # Inputs have to be the same length as labels
    assert len(xs) == len(ys)
    
    # Define the initial parameters to be zero
    p0, p1 = 0, 0
    
    # Find the batch size
    batch_size = len(xs)
    
    for e in range(epochs):
        
        # Perform gradient descent on p0
        sum_p0 = 0.0
        for i in range(batch_size):
            sum_p0 += (p1 * xs[i] + p0) - ys[i]
        sum_p0 /= batch_size
        
        # Perform gradient descent on p1
        sum_p1 = 0.0
        for i in range(batch_size):
            sum_p1 += ((p1 * xs[i] + p0) - ys[i]) * xs[i] 
        sum_p1 /= batch_size
        
        # Calculate the updated parameters
        tmp_p0 = p0 - (lr * sum_p0)
        tmp_p1 = p1 - (lr * sum_p1)
        
        # Update the actual parameters
        p0 = tmp_p0
        p1 = tmp_p1
        
    return (p0, p1)
```

```python
p0, p1 = linear_regression(xs, ys, lr = 0.1, epochs = 4)
```

```python
plt.scatter(xs, ys, c = "red")
plt.xlabel('Years in the education system')
plt.ylabel('Percentage of mental frustration')
plt.axis([0, 4, 0, 1])
plt.plot([0, 4], [get_y_by_slope(p1, p0, 0), get_y_by_slope(p1, p0, 4)])
plt.show()
```
